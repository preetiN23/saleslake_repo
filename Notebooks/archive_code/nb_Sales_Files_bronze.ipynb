```python
dfMonthlySales = spark.read.csv("/Volumes/intellibi_catlog/intellibi_schema/intellibi_data_engineering/src_files/monthly_sales/", header=True, inferSchema=True)
dfMonthlySales.display()
```

```python
%sql
-- CREATE SCHEMA intellibi_catlog.intellibi_bronze;
-- CREATE SCHEMA intellibi_catlog.intellibi_silver;
-- CREATE SCHEMA intellibi_catlog.intellibi_gold;
```

```python
from pyspark.sql import functions as F
dfMonthlySalesClean = dfMonthlySales.selectExpr("UPPER(TRIM(sale_id)) as sale_id", "UPPER(TRIM(product)) as product",
                          "UPPER(TRIM(category)) as category", "TRIM(quantity) as quantity", "TRIM(price) as price",
                          "TRIM(sale_date) as sale_date", "UPPER(TRIM(region)) as region","current_timestamp() as ingest_ts"
                          )
dfMonthlySalesClean.display()
# dfMonthlySalesClean.write.mode("overwrite").saveAsTable("intellibi_catlog.intellibi_bronze.fMonthlySales")                         
#['sale_id', 'product', 'category', 'quantity', 'price', 'sale_date', 'region']
```
