```python
%sql
MERGE INTO intellibi_catlog.intellibi_gold.fMonthlySales
USING intellibi_catlog.intellibi_silver.fMonthlySales
ON intellibi_catlog.intellibi_gold.fMonthlySales.sale_id = intellibi_catlog.intellibi_silver.fMonthlySales.sale_id
WHEN MATCHED THEN UPDATE SET *  
WHEN NOT MATCHED THEN INSERT *
;
```
